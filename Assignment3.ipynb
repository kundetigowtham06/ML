```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##1. Basic Line Plot

*    Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
* Label the x-axis as "X Values" and the y-axis as "Squared Values."
* Add a title to the graph.
  



```python
x = np.arange(-10, 11)
y = x**2
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('X^2')
plt.title('Parabola')
plt.show()
```

##2. Bar Plot

* Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
* Add labels to the x-axis and y-axis and give the chart a title.   



```python
x=['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
y=[85, 90, 78, 92, 88]
plt.bar(x, y)
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show()
```

##3. Scatter Plot

*   Generate two lists of 20 random integers each (for x and y coordinates).

size, and labels to the points.

##4. Histogram


*   Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
* Customize the number of bins and add a grid to the plot.




```python
data=np.random.randn(101)
plt.hist(data, bins=10)
plt.hist(data,bins=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Histogram")
plt.grid()
plt.show()
```

##5. Pie Chart

*    Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
* Display the percentage values on the chart and use a color palette for
differentiation.
   



```python
d=[20,30,25,15,10]
i=['A','B','C','D','E']
plt.pie(d,labels=i,autopct='%1.1d%%')
plt.show()
```

##6. Box Plot (Seaborn)

students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].

##7. Heatmap (Seaborn)

*    Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.
* Add a color bar and annotate the heatmap with the values.  



```python
m=np.random.randint(1,11, size=(5,5))
sns.heatmap(m,annot=True,cmap='YlGnBu')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Heat-map")
plt.show()
```
