```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/world_health_data.csv')
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['health_exp'].fillna(data['health_exp'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['life_expect'].fillna(data['life_expect'].mean(),inplace=True)
data['maternal_mortality'].fillna(data['maternal_mortality'].mean(),inplace=True)
data['infant_mortality'].fillna(data['infant_mortality'].mean(),inplace=True)
data['neonatal_mortality'].fillna(data['neonatal_mortality'].mean(),inplace=True)
data['under_5_mortality'].fillna(data['under_5_mortality'].mean(),inplace=True)
data['prev_hiv'].fillna(data['prev_hiv'].mean(),inplace=True)
data['inci_tuberc'].fillna(data['inci_tuberc'].mean(),inplace=True)
data['prev_undernourishment'].fillna(data['prev_undernourishment'].mean(),inplace=True)



```

```python
data.isnull().sum()
```

```python
data.sample(10)
```

```python
numerical_data = data.select_dtypes(include=np.number)
m=numerical_data.corr().abs()
plt.figure(figsize=(8, 6))
sns.heatmap(m, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['health_exp'], kde=True, color='blue', bins=30)
plt.title('Distribution of Health Expenditure (% of GDP)')
plt.xlabel('Health Expenditure (% of GDP)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='maternal_mortality', y='infant_mortality', data=data, color='blue', alpha=0.7)
plt.title('Infant Mortality vs Maternal Mortality')
plt.xlabel('Maternal Mortality (per 100,000 live births)')
plt.ylabel('Infant Mortality (per 1,000 live births)')
plt.show()
```
