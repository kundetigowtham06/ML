##. Data Transformation
 Load a dataset containing daily temperatures and dates. Perform the following:

*    Convert the date column into a DateTime object.
* Create a new column that categorizes temperatures into High, Medium, or
Low.
  



```python
data=pd.read_csv("/content/DailyDelhiClimateTest.csv")
data['date']=pd.to_datetime(data['date'])
data
```

```python
#Create a new column that categorizes temperatures into High, Medium, or Low.
def temperature(temp):
  if temp>=25:
    return 'High'
  elif temp>=20:
    return 'Medium'
  else:
    return 'Low'
data['Temperature']=data['meantemp'].apply(temperature)
data
```
