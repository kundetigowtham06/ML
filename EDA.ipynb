#100 Richest People In World


This dataset contains the list of Top 100 Richest People in the World

* Name - Person Name
* NetWorth - His/Her Networth
* Age - Person Age
* Country - The country person belongs to
* Source - Information Source
* Industry - Expertise Domain

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('/content/TopRichestInWorld.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

There are no columns having null values.

```python
data.info()
```

```python
data["Source"].value_counts()
```

```python
data.describe()
```

```python
def remove_comma(x):
    if "," in x:
        s=x.replace("," , "")
        return s
    else:
        return x
```

```python
data["NetWorth"]=data["NetWorth"].apply(remove_comma)
data["NetWorth"]=data["NetWorth"].str.replace("$","").astype(int)
```

##**Top 5 Richest persons**

```python
top=data.head()
top
```

```python
sns.barplot(x=top["Name"],y=top["NetWorth"])
plt.xticks(rotation=40)
plt.figure(figsize=(15,8))
plt.show()
```

#Country having Highest richest persons

```python
data=data.rename(columns={"Country/Territory":"Country"})
```

```python
sns.countplot(x=data["Country"].sort_values() ,order=data["Country"].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

#In which Industry they earning profit more?

```python
sns.countplot(x=data["Industry"].sort_values() ,order=data["Industry"].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

#Age Distribution

```python
bins = [30,40,50,60,70,80,90,100]

plt.hist(data.Age, bins = bins, edgecolor='white')

plt.ylabel('Number of People')
plt.xlabel('Age')
plt.title('Age Distribution of the 100 Richest People Worldwide')
```
